## matplotlib

### Let's create a simple plot!

 In this section, we want to draw the cosine and sine functions on the same plot. 

Create input data

```python
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 30, endpoint=True)
```

```python
c, s = np.cos(x), np.sin(x)
```

Plot the data with default settings

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot as a line
plt.plot(x, c)
plt.plot(x, s)
```

```python
# Plot as dots
plt.scatter(x,c)
plt.scatter(x,s)
```

### Modify the plot

Change colours and line widths

```python
# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(x, c, color="blue", linewidth=1.0, linestyle="-")

# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(x, s, color="green", linewidth=1.0, linestyle="-")
```

Set limits

```python
# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(x, c, color="blue", linewidth=1.0, linestyle="-")

# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(x, s, color="green", linewidth=1.0, linestyle="-")

# Set x limits
plt.xlim(-4.0, 4.0)

# Set y limits
plt.ylim(-1.0, 1.0)
```

Add a legend

```python
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(x, s, color="red",  linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')
```

Plot two lines as separate subplots

```python
fig = plt.figure()

plt.subplot(1, 2, 1)
plt.plot(x, c, color="blue")

plt.subplot(1, 2, 2)
plt.plot(x, s, color="red")
```
